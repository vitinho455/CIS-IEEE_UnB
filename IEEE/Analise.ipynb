```python

import seaborn as sns
import matplotlib.pyplot as plt

table = pd.read_csv('diabetes.csv')

tabelas = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']

# Representação da distribuição em Histograma

for coluna in tabelas: 

    plt.figure(figsize=(8, 4))
    sns.histplot(table[coluna], kde=True)
    plt.title("Histogram")
    plt.xlabel(coluna)
    plt.ylabel("Frequency")
    plt.show()
```

# Questão 5

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

table = pd.read_csv('diabetes.csv')

# Média do IMC do grupo 0
bmi_mean_zero = table[table['Outcome'] == 0]['BMI'].mean()

print(f"Media do IMC no grupo 0 : {bmi_mean_zero: .2f}")

# Média do IMC do grupo 1
bmi_mean_one = table[table['Outcome'] == 1]['BMI'].mean()

print(f"Media do IMC no grupo 1 : {bmi_mean_one: .2f}")

# Representação no gráfico boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(data=table, x='Outcome', y='BMI')
plt.title('Boxplot')
plt.xlabel('Outcome')
plt.ylabel('BMI')
plt.show()
```

# Questão 6

```python
import numpy as np

table = pd.read_csv('diabetes.csv')

#Valor considerado crítico da glicose.  (Glicose >= 180 é considerado Hiperglicemia em pacientes críticos)

critical_glucose_value = table[(table['Glucose'] >= 180.0) & table['Outcome'] == 1]

print(critical_glucose_value.to_string())

# Correlação 
critical_glucose_value.corr()
```

# Questão 9

```python
import matplotlib.pyplot as plt

table = pd.read_csv('diabetes.csv')

# Pessoas com idade maior ou igual a 50 e que são diabéticos
age_fifty = table[(table['Age'] >= 50) & table['Outcome'] == 1].shape[0]

# Pessoas com idade menor ou igual a 50 e que são diabéticos
young_age = table[(table['Age'] <= 50) & table['Outcome'] == 1].shape[0]

categories = ['Idade >= 50 e Outcome = 1', 'Idade < 50 e Outcome = 1']
values = [age_fifty, young_age]

# Gráfico de Barras
plt.bar(categories, values)

plt.title('Taxas de Diabetes')
plt.xlabel('Categorias')
plt.ylabel('Quantidade')

plt.show()
```

# Conclusões

- Questão 1: Foram encontrados valores faltantes em variáveis que exigem outras informações para serem preenchidas corretamente. Um exemplo é o IMC (calculado com base no peso e altura). Com isso, esses valores foram tratados como NaN (Not a Number), indicando dados ausentes.

- Questão 2: As variáveis estão, em geral, bem distribuídas. Porém, alguns valores destoam e geram confusão, como por exemplo DiabetesPedigreeFunction = 627.00, que parece fora de escala e pode indicar erro ou outlier.

- Questão 3: Sim, há correlação. Pessoas mais jovens estão apresentando mais casos de diabetes do que os mais velhos, como evidenciado principalmente no gráfico de barras apresentado na questão 9.

- Questão 4: As variáveis Glucose, Insulin, BMI e DiabetesPedigreeFunction mostraram maior correlação com a presença de diabetes. Com base nessas informações, é possível estimar a chance de um paciente desenvolver a doença, embora não seja suficiente para uma definição conclusiva.

- Questão 5: Sim. Pacientes com IMC maior ou igual a 30 apresentam maior probabilidade de desenvolver diabetes. A média de IMC no grupo sem diabetes foi de 30.38, enquanto no grupo com diabetes foi de 35.02 — ambas consideradas elevadas, indicando alto risco para os dois grupos.

- Questão 6: O valor crítico para a glicose pode ser considerado 180 mg/dL ou mais, o que representa um estado de hiperglicemia e risco elevado à saúde. As taxas nessa faixa são alarmantes e indicam complicações sérias.

- Questão 8: A variável DiabetesPedigreeFunction faz sentido, já que representa o histórico familiar da doença. Apesar de alguns valores parecerem estranhos, a ideia de que a linhagem genética influencia no risco de diabetes é coerente.

- Questão 9: O gráfico revelou que pessoas mais jovens estão sendo mais afetadas pela diabetes do que as mais velhas — um resultado surpreendente. Fatores como IMC elevado e níveis altos de glicose podem estar contribuindo para esse aumento expressivo de casos entre jovens.
