```python
import pandas as pd 
import numpy as np

table = pd.read_csv('diabetes.csv')

# Substituio dos valores 0 e " " para NaN

columns = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']

table[columns] = table[columns].replace(0, np.nan)

print(table.to_string())
```

```python

import seaborn as sns
import matplotlib.pyplot as plt

table = pd.read_csv('diabetes.csv')

tabelas = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']

# Representao da distribuio em Histograma

for coluna in tabelas: 

    plt.figure(figsize=(8, 4))
    sns.histplot(table[coluna], kde=True)
    plt.title("Histogram")
    plt.xlabel(coluna)
    plt.ylabel("Frequency")
    plt.show()
```

# Questo 5

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

table = pd.read_csv('diabetes.csv')

# Mdia do IMC do grupo 0
bmi_mean_zero = table[table['Outcome'] == 0]['BMI'].mean()

print(f"Media do IMC no grupo 0 : {bmi_mean_zero: .2f}")

# Mdia do IMC do grupo 1
bmi_mean_one = table[table['Outcome'] == 1]['BMI'].mean()

print(f"Media do IMC no grupo 1 : {bmi_mean_one: .2f}")

# Representao no grfico boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(data=table, x='Outcome', y='BMI')
plt.title('Boxplot')
plt.xlabel('Outcome')
plt.ylabel('BMI')
plt.show()
```

```python
import numpy as np

table = pd.read_csv('diabetes.csv')

#Valor considerado crtico da glicose.  (Glicose >= 180  considerado Hiperglicemia em pacientes crticos)

critical_glucose_value = table[(table['Glucose'] >= 180.0) & table['Outcome'] == 1]

print(critical_glucose_value.to_string())

# Correlao 
critical_glucose_value.corr()
```

# Questo 9

```python
import matplotlib.pyplot as plt

table = pd.read_csv('diabetes.csv')

# Pessoas com idade maior ou igual a 50 e que so diabticos
age_fifty = table[(table['Age'] >= 50) & table['Outcome'] == 1].shape[0]

# Pessoas com idade menor ou igual a 50 e que so diabticos
young_age = table[(table['Age'] <= 50) & table['Outcome'] == 1].shape[0]

categories = ['Idade >= 50 e Outcome = 1', 'Idade < 50 e Outcome = 1']
values = [age_fifty, young_age]

# Grfico de Barras
plt.bar(categories, values)

plt.title('Taxas de Diabetes')
plt.xlabel('Categorias')
plt.ylabel('Quantidade')

plt.show()
```
